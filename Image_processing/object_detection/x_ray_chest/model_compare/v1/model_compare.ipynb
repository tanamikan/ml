```python
import matplotlib.pyplot as plt
import numpy as np

evaluate_list1 = [66, 50, 32]
evaluate_list2 = [0, 16, 16]

plt.rcParams['font.family']= 'sans-serif'
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.width'] = 1.2
plt.rcParams['ytick.major.width'] = 1.2
plt.rcParams['font.size'] = 12
plt.rcParams['axes.linewidth'] = 1.2
plt.rcParams['axes.grid']=True
plt.rcParams['grid.linestyle']='--'
plt.rcParams['grid.linewidth'] = 0.3
plt.rcParams["legend.markerscale"] = 2
plt.rcParams["legend.fancybox"] = False
plt.rcParams["legend.framealpha"] = 1
plt.rcParams["legend.edgecolor"] = 'black'

fig, ax = plt.subplots(1, 1, figsize=(6, 4.5))

left = np.arange(len(evaluate_list1))
width = 0.3

labels = ['Faster-R-CNN', 'SSD', 'YOLOv8']

ax.bar(left, evaluate_list1, color='red', width=width, align='center', label='Pattern1')
ax.bar(left+width, evaluate_list2, color='blue', width=width, align='center', label='Pattern2')

ax.set_xlabel('Model')
ax.set_ylabel('Match percentage')

plt.xticks(left + width/2, labels)
plt.legend()

plt.show()
```
