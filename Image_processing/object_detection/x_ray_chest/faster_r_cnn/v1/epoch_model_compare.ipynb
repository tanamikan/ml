```python
import os
from glob import glob
import numpy as np
import pandas as pd
import seaborn as sns

base_path = '/content/drive/MyDrive/x_ray_chest/results/faster_r_cnn'
ver = 'v1'

# 新しいフォルダを生成
weight_folder = 'weight'
train_folder = 'train'
test_folder = 'test'
val_folder = 'val'
test_img_folder = 'images'

weight_folder_path = os.path.join(base_path, ver, weight_folder)
train_folder_path = os.path.join(base_path, ver, train_folder)
test_folder_path = os.path.join(base_path, ver, test_folder)
val_folder_path = os.path.join(base_path, ver, val_folder)
test_img_path = os.path.join(base_path, ver, test_folder, test_img_folder)
```

```python
loss_outdir = os.path.join(train_folder_path, 'train_loss.csv')
loss_df = pd.read_csv(loss_outdir)

val_loss_index = loss_df.nsmallest(50, 'val loss')
val_loss_index_list = val_loss_index['epoch'].tolist()

print(val_loss_index_list)
```

```python
map_csv_path = os.path.join(val_folder_path, 'map.csv')
map_result_df = pd.read_csv(map_csv_path)

map_val_index = map_result_df.nlargest(50, 'map')
map_val_index_list = map_val_index['epoch'].tolist()

print(map_val_index_list)
```

```python
map_csv_path = os.path.join(test_folder_path, 'map.csv')
map_result_df = pd.read_csv(map_csv_path)

map_test_index = map_result_df.nlargest(50, 'map')
map_test_index_list = map_test_index['epoch'].tolist()

print(map_test_index_list)
```

```python
match_count = sum(1 for elem in map_test_index_list if elem in map_val_index_list)

match_percentage = (match_count / len(map_test_index_list)) * 100

print(match_percentage)
```

```python
match_count = sum(1 for elem in map_test_index_list if elem in val_loss_index_list)

match_percentage = (match_count / len(map_test_index_list)) * 100

print(match_percentage)
```
