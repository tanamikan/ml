```python
from ultralytics.models import YOLO
import cv2
from glob import glob
import os
```

```python
weight_dirs = glob('/content/drive/MyDrive/x_ray_chest/results/yolo/v1/weight/*')

for weight_dir in weight_dirs:

    weight_name = str(os.path.basename(weight_dir).split('.')[0])

    # Load a model
    model = YOLO(weight_dir)

    # Run inference on the source
    results = model.predict(source = '/content/drive/MyDrive/x_ray_chest/results/yolo/dataset/v1/images/test/*',
                            project = '/content/drive/MyDrive/x_ray_chest/results/yolo/v1/test/predict',
                            name = weight_name,
                            conf=0.1,
                            iou=0.5,
                            save_txt=True,
                            save_conf=True
                            )
```
