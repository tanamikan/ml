```python
import pandas as pd
import os

loss_file_path = '/content/drive/MyDrive/x_ray_chest/results/yolo/v1/weight/results.csv'

loss_df = pd.read_csv(loss_file_path, encoding='utf-8')

loss_df.replace('                    nan', '', inplace=True)

print(loss_df.head())
```

```python
loss_df['train_loss_sum'] = loss_df.iloc[:, 1] + loss_df.iloc[:, 2] + loss_df.iloc[:, 3]
loss_df['val_loss_sum'] = pd.to_numeric(loss_df.iloc[:, 8]) + pd.to_numeric(loss_df.iloc[:, 9]) + pd.to_numeric(loss_df.iloc[:, 10])

loss_df.head()
```

```python
min_val_value = loss_df['val_loss_sum'].min()

min_index = loss_df.index[loss_df['val_loss_sum'] == min_val_value]
min_epoch = loss_df.iloc[min_index, 0].values[0]

print(f'min_epoch: {min_epoch}, min_val_value: {min_val_value}')
```

```python
from matplotlib import pyplot as plt
import numpy as np

# ÂÖ®‰ΩìË®≠ÂÆö
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rcParams['xtick.major.width'] = 1.2
plt.rcParams['ytick.major.width'] = 1.2
plt.rcParams['font.size'] = 12
plt.rcParams['axes.linewidth'] = 1.2
plt.rcParams['axes.grid']=True
plt.rcParams['grid.linestyle']='--'
plt.rcParams['grid.linewidth'] = 0.3
plt.rcParams["legend.markerscale"] = 2
plt.rcParams["legend.fancybox"] = False
plt.rcParams["legend.framealpha"] = 1
plt.rcParams["legend.edgecolor"] = 'black'

fig = plt.figure()

ax1 = fig.subplots()
ax2 = ax1.twinx()

ax1.plot(loss_df.iloc[:, 0], loss_df['train_loss_sum'], color="blue", label="train loss")
ax2.plot(loss_df.iloc[:, 0], loss_df['val_loss_sum'], c="r", label="val loss")

ax1.set_xlim(0, 200)
ax1.set_ylim(0, 14)
ax1.set_xlabel('epochs')
ax1.set_ylabel('train loss')

ax2.set_xlim(0, 200)
ax2.set_ylim(0, 14)
ax2.set_xlabel('epochs')
ax2.set_ylabel('val loss')

plt.grid(color = "gray", linestyle="dotted")

plt.title("train loss, val loss vs epochs")
h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2)

plt.show()
```
